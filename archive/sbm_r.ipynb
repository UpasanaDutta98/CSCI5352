```r
library(blockmodels)
library(igraph)
```

```r
g <- read_graph(file = "demograph.gml", format = "gml")
```

```r
g
```

```r
g_adj <- as.matrix(as_adj(g, type = "both", attr = NULL, edges = FALSE, names = TRUE))
```

```r
bm <- BM_gaussian(membership_type = "SBM", adj = g_adj, verbosity=6, autosave='', plotting=character(0), 
                  exploration_factor=1.5, explore_min=2, explore_max=6, ncores=detectCores())
```

```r
bm$estimate()
```

```r
which.max(bm$ICL)
```
